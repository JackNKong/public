```python
import pandas as pd
df = pd.read_csv('./과제물/ThoraricSurgery.csv',header= None)

df
```

```python
data_set = df.values
X= data_set[:,:17]
y_cls = data_set [:,17]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
e = LabelEncoder()
e.fit(y_cls)
y = e.transform(y_cls)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
import numpy as np

X_tensor = torch.tensor(X.astype(float)).float()
y_tensor = torch.tensor(y.astype(np.int64))
```

```python
model = nn.Sequential(
    nn.Linear(17,68),
    nn.ReLU(),
    nn.Linear(68,34),
    nn.ReLU(),
    nn.Linear(34,3),
    #nn.Sigmoid()
)
```

```python
loss_func = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
#loss_list = []
#output=[i[0]for i in output]

for i in range(3000) : #시간상 100 원래 3000
    optimizer.zero_grad()
    output = model(X_tensor)
    #output = output.view([470])
    #output = netD(real_cpu).view(-1)
    #output = output.type(torch.FloatTensor)
    loss = loss_func(output, y_tensor)
    loss.backward()
    optimizer.step()
    if i%100 == 0 : #시간상 5개 원래 100
        print(loss.item())
       # loss_list.append(loss.item())
```

```python
output = model(X_tensor)
```

```python
pred= torch.argmax(output, dim=1)
pred_np = pred.detach().numpy()
accu = np.mean(pred_np==y)
print(accu)
```

```python
sm = nn.Softmax(dim=1)
prob=sm(output)
```
